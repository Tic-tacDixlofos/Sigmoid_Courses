Probability is a measure of uncertainty. Probability applies to machine
learning because in the real world, we need to make decisions with
incomplete information


• Frequentist probability - considers the actual likelihood of an
event;

• Bayesian probability - considers how strongly we believe that an
event will occur.


Naive Bayes (наивный байесовский классификатор) — это алгоритм машинного обучения, основанный на теореме Байеса с наивным (очень наивным!) предположением:
все признаки независимы друг от друга, при условии класса. Это значит, что мы должны обработать признаки так, чтобыони считаллись независимыми ург от друга


Наивный Байес:

игнорирует корреляцию полностью

перемножает вероятности признаков, как будто они не зависят друг от друга

может переоценивать или недооценивать вероятности, потому что использует повторно “похожую” информацию

```python
import kagglehub
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from collections import OrderedDict
from functools import reduce
from operator import mul
from sklearn.naive_bayes import GaussianNB

from sklearn.preprocessing import OneHotEncoder
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, Binarizer, OneHotEncoder
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score
```

```python
path = kagglehub.dataset_download("blastchar/telco-customer-churn")

# Загружаем CSV
csv_path = os.path.join(path, "WA_Fn-UseC_-Telco-Customer-Churn.csv")
dt = pd.read_csv(csv_path)

# Отображаем первые строки
print(dt.head())
```

```python
dt
```

Split the data set into the train and test set.

```python
X = dt.drop("Churn", axis=1)
Y = dt['Churn']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

Data Analyse

```python
dt.describe()
```

```python
numerical_cols = X_train.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = X_train.select_dtypes(include=['object', 'string']).columns
```

```python
scaler = StandardScaler()
X_train_num_scaled = scaler.fit_transform(X_train[numerical_cols])
X_test_num_scaled = scaler.transform(X_test[numerical_cols])
```

```python
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
X_train_cat_encoded = ohe.fit_transform(X_train[categorical_cols])
X_test_cat_encoded = ohe.transform(X_test[categorical_cols])
```

Unite

```python
X_train_processed = np.hstack([X_train_num_scaled, X_train_cat_encoded])
X_test_processed = np.hstack([X_test_num_scaled, X_test_cat_encoded])
```

#Создаём версии данных для каждой модели

```python
X_train_gaussian = X_train_processed
X_test_gaussian = X_test_processed
```

Для MultinomialNB — данные должны быть неотрицательными

```python
min_val = X_train_processed.min()
X_train_multinomial = X_train_processed - min_val
X_test_multinomial = X_test_processed - min_val
```

Для BernoulliNB — бинаризация

```python
binarizer = Binarizer()
X_train_bernoulli = binarizer.fit_transform(X_train_gaussian)
X_test_bernoulli = binarizer.transform(X_test_gaussian)
```

Обучаем модели и считаем accuracy

```python
results = []
```

# MODEL TRAININGS

Для GaussianNB — стандартизированные данные

Create an instance of Gaussian Naive Bayes

```python
gnb = GaussianNB()
gnb.fit(X_train_gaussian, Y_train)
gnb_pred = gnb.predict(X_test_gaussian)
results.append(['GaussianNB', accuracy_score(Y_test, gnb_pred)])
```

Create an instance of Multinomial Naive Bayes

```python
mnb = MultinomialNB()
mnb.fit(X_train_multinomial, Y_train)
mnb_pred = mnb.predict(X_test_multinomial)
results.append(['MultinomialNB', accuracy_score(Y_test, mnb_pred)])
```

Create an instance of Bernoulli Naive Bayes

```python
bnb = BernoulliNB()
bnb.fit(X_train_bernoulli, Y_train)
bnb_pred = bnb.predict(X_test_bernoulli)
results.append(['BernoulliNB', accuracy_score(Y_test, bnb_pred)])
```

Create a table of the following structure as a pandas data frame.

```python
summary_df = pd.DataFrame(results, columns=['Model', 'Accuracy'])
print(summary_df)
```
