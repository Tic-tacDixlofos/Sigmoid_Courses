```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uciml/pima-indians-diabetes-database")

print("Path to dataset files:", path)
```

```python
from pathlib import Path
import pandas as pd
csv_files = list(Path(path).rglob("*.csv"))
df = pd.read_csv(csv_files[0])
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

MATLAB-style Interface

```python
plt.plot(df['Age'], df['Glucose'])
plt.xlabel("Age")
plt.ylabel("Glucose")
plt.title("Glucose vs Age")
plt.grid(True)
plt.show()
```

Object Oriented Interface

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(df['Age'], df['Glucose'], color='teal')
ax.set_title("Glucose vs Age (OOP)")
ax.set_xlabel("Age")
ax.set_ylabel("Glucose")
ax.grid(True)
plt.show()
```

Simple Line Plot

```python
plt.plot(df['Glucose'].head(20), label="Glucose")
plt.plot(df['BMI'].head(20), label="BMI")
plt.legend()
plt.title("Line Plot")
plt.show()
```

Simple Scatter Plot

```python
plt.scatter(df['Glucose'], df['BMI'], alpha=0.6)
plt.xlabel("Glucose")
plt.ylabel("BMI")
plt.title("Scatter: Glucose vs BMI")
plt.show()
```

Error Bars

```python
import numpy as np

means = df.groupby("Outcome")["Glucose"].mean()
stds = df.groupby("Outcome")["Glucose"].std()
outcomes = [0, 1]

plt.errorbar(outcomes, means, yerr=stds, fmt='o', capsize=5, color='purple')
plt.xticks(outcomes, ['No Diabetes', 'Diabetes'])
plt.ylabel("Glucose")
plt.title("Error Bars for Glucose by Outcome")
plt.show()
```

Contour Plot

```python
from scipy.stats import gaussian_kde

x = df["Glucose"]
y = df["BMI"]
xy = np.vstack([x, y])
z = gaussian_kde(xy)(xy)

plt.figure(figsize=(6, 4))
plt.tricontourf(x, y, z, levels=15, cmap="plasma")
plt.colorbar()
plt.title("Contour Plot: Glucose vs BMI")
plt.xlabel("Glucose")
plt.ylabel("BMI")
plt.show()
```

Histograms

```python
plt.hist(df["Age"], bins=20, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

KDE Plot with seaborn

```python
import seaborn as sns
sns.kdeplot(df["Insulin"], fill=True, color="red")
plt.title("Insulin Density")
plt.show()
```

Jointplot

```python
sns.jointplot(data=df, x="Glucose", y="BMI", kind="scatter", hue="Outcome")
```

Boxplot

```python
sns.boxplot(data=df, x="Outcome", y="Glucose")
plt.title("Glucose by Diabetes Outcome")
```

Violinplot

```python
sns.violinplot(data=df, x="Outcome", y="BMI", palette="muted")
plt.title("BMI Distribution by Outcome")
```

lmplot  /

```python
sns.lmplot(data=df, x="Age", y="Glucose", hue="Outcome", height=5, aspect=1.3)
```
