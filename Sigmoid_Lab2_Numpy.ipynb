# Sigmoid


The function sigmoid(x) turns any number (x) into a value between 0 and 1.

It is like translating a “normal number” into a “probability” (0% to 100%).

```python
def sigmoid(x):
  return  1 / (1 + np.exp(-x))

sigmoid(x)
```

Используется, чтобы превратить результат модели в вероятность. Например: вероятность, что на фото кошка — 0.9 (или 90%). Очень полезно в задачах, где нужно сказать «да» или «нет» (классификация).


Результаты показывают как бы «уверенность» модели. Например, если использовать в машинном обучении, это можно читать так:

0.61 → «61% уверенности»

0.67 → «67% уверенности»


Все твои x находятся в такой зоне, где сигмоида выдаёт значения больше 0.5 — значит, твои x в основном положительные и не слишком маленькие.

Все твои результаты — это теперь как бы «вероятности» или «уверенность» в каком-то предсказании, если бы ты использовала это в модели (например, предсказать «да» или «нет»).

# Weights Update in Logistic Regression


To train the model. The model predicts poorly at first (e.g., spam or not), and we use this formula to “move” the weights a bit so that it predicts better next time.

```python
# Делаем бинарные метки: хорошее вино (1) если quality >= 6, иначе плохое (0)
y = (dt['quality'].values >= 6).astype(int)

# Возьмем два признака: алкоголь и плотность (density)
X = dt[['alcohol', 'density']].values

# Добавим столбец 1 для bias (сдвига)
X = np.hstack([np.ones((X.shape[0], 1)), X])

# Инициализируем веса нулями
w = np.zeros(X.shape[1])

# Шаг обучения
alpha = 0.01

# Предсказание
pred = sigmoid(X @ w)

# Градиент
gradient = X.T @ (pred - y)

# Обновление весов
w = w - alpha * gradient

print("Обновлённые веса:", w)
```

Я задалась вопросом, что такое веса, это как сильно факторы влияют на результат. Допустим модель решает, какое вино качественное или нет, она смотрит на факторы и дает результат, как мы видим по фенкции симоила возможно правильный около 60 процентов.

Если модель ошибается она будет уменьгать или увеличивать фанктор веса, то есть, если модель опиралась на то, что тамый важный фактор хлориды например, модель ошибается и делает вывод, что хлориды не самый ваэный фактор и его вес немго уменьшается в знаечении.

А с кодом я работаю так:

Ограничила предсказания, чтобы не сломать log.

Убираем ноль.


Посчитала, насколько модель уверенно угадывает 0 и 1.

Если угадывает уверенно но неправльно плохо.

Если угадывает неуверенно с меньшей вероятностью, то хорошо.

Посчитала среднюю ошибку — вот это и есть Binary Cross Entropy.



# MSE

Let's calculate the error between the predicted probabilities and the real labels

Чем меньше это число, тем точнее моя модель.

```python
mse = np.mean((y - pred) ** 2)
print("Mean Squared Error:", mse)
```

# Binary Cross Entropy

Let's calculate a more appropriate error for binary classification (yes/no).

Что это:
Это ещё один способ измерить ошибку, но именно для задач, где нужно выбрать между двумя вариантами (например: «спам» или «не спам»).

Зачем это нужно:
Она помогает правильно наказывать модель за неправильные предсказания вероятностей и поощрять за правильные. Благодаря этому модель учится давать уверенные и точные ответы.


Если модель говорит 0.99 (очень уверенно), а на самом деле это был 0 (неправда) → ошибка будет огромной.

А если модель сказала 0.51 (почти наугад) — и ошиблась → её «накажут» поменьше.

То есть бинарная кросс-энтропия как строгий учитель:

Сильно ругает за уверенные, но неправильные ответы.

И мягче относится, если модель была неуверенна.

```python
# Чтобы не было log(0), ограничим значения
eps = 1e-15
pred = np.clip(pred, eps, 1 - eps)

bce = -np.mean(y * np.log(pred) + (1 - y) * np.log(1 - pred))
print("Binary Cross Entropy:", bce)
```
